```python
import pandas as pd
import pickle
import numpy as np
```

```python
X_train = pickle.load(open("tfidf_X_tr","rb"))
X_test = pickle.load(open("tfidf_X_test","rb"))
y_train=pd.read_csv('train_y.csv')
y_test=pd.read_csv('test_y.csv')

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logModel =  LogisticRegression()
```

```python
param_grid = [    
    {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-4, 4, 5),
    'solver' : ['sag','saga'],
    'max_iter' : [100, 200],
     'verbose' : [1],
     'multi_class': ['auto']
    }
]
```

```python
from sklearn.model_selection import GridSearchCV
logModel_tuned = GridSearchCV(logModel, param_grid, cv = 5, scoring='accuracy')
logModel_tuned.fit(X_train, y_train)
```

```python
print(logModel_tuned.best_params_)
```

```python
y_pred = logModel_tuned.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
